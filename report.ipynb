### Сравнение policy и value iteration по скорости сходимости и работы

```python
time_result = pd.DataFrame()
for value_th in [0.01, 0.001, 0.0001]:
    for env_size in ['small', 'medium', 'large', 'huge']:
        for g in [0.1, 0.5]:
            print(env_size)
            env = gym.make('frozen_lake:default-v0', map_name= env_size, action_set_name='slippery')
            env.reset()
            alg = PolicyIteration(env, gamma=g, eval_policy_th=eval_policy_th)
            value_times = []
            value_iters = []
            for e in range(10):
                start_time = time.time()
                p, v, i_value = alg.value_iteration(value_th)
                value_time = time.time() - start_time
                value_times.append(value_time)
                value_iters.append(i_value)
            i_value = np.mean(value_iters)
            value_time = np.mean(value_times)
            policy_times = []
            policy_iters = []
            for e in range(10):
                start_time = time.time()
                p, v, i_policy = alg.policy_iteration()
                policy_time = time.time() - start_time
                policy_times.append(policy_time)
                policy_iters.append(i_policy)
            i_policy = np.mean(policy_iters)
            policy_time = np.mean(policy_times)

            time_result = time_result.append({'env_size': env_size, f'value_time_gamma={g}_th={value_th}': value_time,
                                              f'value_iter_gamma={g}_th={value_th}': i_value, f'policy_time_gamma={g}': policy_time,
                                              f'policy_iter_gamma={g}': i_policy}, ignore_index=True)

time_result.set_index('env_size', inplace=True)
```

```python
temp = pd.pivot_table(time_result, index=['env_size'])
```

```python
temp.loc[:, [col for col in time_result.columns if 'time' in col]].plot(title='Время исполнения, ms', figsize=(20, 10))
```

```python
temp.loc[:, [col for col in time_result.columns if 'iter' in col]].plot(title='Сходимость, итерации', figsize=(20, 10))
```
